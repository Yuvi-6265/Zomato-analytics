# Basic Statistics: Calculate and visualize the average rating of restaurants.
# Analyze the distribution of restaurant ratings to understand the overall rating landscape.


```python
#Calculate the Average Rating

avg_rating = df['aggregate_rating'].mean()

print("Average Restaurant Rating:", round(avg_rating, 2))
```

```python
#visualize the Distribution of Ratings

plt.figure(figsize=(8,5))

sns.histplot(
    df['aggregate_rating'],
    bins=20,
    kde=True
)

plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Box Plot of Ratings

plt.figure(figsize=(8,4))

sns.boxplot(x=df['aggregate_rating'])

plt.title('Box Plot of Restaurant Ratings')
plt.show()
```

# Top Restaurant Chains: Identify and visualize the top restaurant chains based on the number of outlets. Explore the ratings of these top chains

```python
# top restaurants

top_chains = df['name'].value_counts().head(10)

print(top_chains)
```

```python
# average Ratings of Top Chains

top_chains = df['name'].value_counts().head(10).index

df[df['name'].isin(top_chains)].groupby('name')['aggregate_rating'].mean()
```

```python
# visualization


ratings = df[df['name'].isin(top_chains)].groupby('name')['aggregate_rating'].mean()
ratings.plot(kind='bar')

plt.title('Ratings of Top Restaurant Chains')
plt.xlabel('Restaurant Name')
plt.ylabel('Average Rating')

plt.show()
```
